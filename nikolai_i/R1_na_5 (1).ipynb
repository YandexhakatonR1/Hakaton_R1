```python
def corrFilter(x: pd.DataFrame, bound: float):
    xCorr = x.corr()
    xFiltered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    xFlattened = xFiltered.unstack().sort_values().drop_duplicates()
    return xFlattened
```

```python
df_train, df_test = train_test_split(df_1, test_size=0.01, random_state=42)
```

```python
corrFilter(df_test, .9)
```

```python
df_train, df_test2 = train_test_split(df_1, test_size=0.05, random_state=42)
```

```python
corrFilter(df_test2, .9)
```

```python
dfCorr = df_test2.corr()
filteredDf = dfCorr[((dfCorr >= .5) | (dfCorr <= -.5)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="Reds")
plt.show()
```
